```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a DataFrame from the data
df = pd.read_excel('Citizen science v literature.xlsx')
```

```python
# Group the data by 'type' and 'year' and count the occurrences
grouped_data = df.groupby(['type', 'year']).size().reset_index(name='count')

# Pivot the data to have 'type' as columns and 'year' as the index
pivot_data = grouped_data.pivot(index='year', columns='type', values='count').fillna(0)

# Create the dual-axis chart with line graphs
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

pivot_data['citizen science'].plot(kind='line', color='b', ax=ax1, label='Citizen Science')
pivot_data['literature'].plot(kind='line', color='g', ax=ax2, label='Literature')

ax1.set_ylabel('Count (Citizen Science)', color='b')
ax2.set_ylabel('Count (Literature)', color='g')

ax1.set_xlabel('Year')
ax1.set_title('Dual-Axis Chart with Line Graphs')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set()



# Pivot the data to have 'type' as columns and 'year' as the index
pivot_data = grouped_data.pivot(index='year', columns='type', values='count').fillna(0)

# Ensure the DataFrame includes all years from the earliest to the latest
earliest_year = min(pivot_data.index)
latest_year = max(pivot_data.index)
all_years = range(earliest_year, latest_year + 1)
pivot_data = pivot_data.reindex(all_years, fill_value=0)

# Create two separate subplots, one for 'Citizen Science' and one for 'Literature'
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), sharex=True)

# Plot the line graph for 'Citizen Science' on the upper subplot
ax0 = pivot_data['citizen science'].plot(kind='line', ax=axes[0], label='Citizen Science', color='b')
ax0.set_ylabel('Count')
ax0.set_title('Upper Subplot - Citizen Science Line Graph')
ax0.legend()

# Plot the line graph for 'Literature' on the lower subplot
ax1 = pivot_data['literature'].plot(kind='line', ax=axes[1], label='Literature', color='g')
ax1.set_ylabel('Count')
ax1.set_xlabel('Year')
ax1.set_title('Lower Subplot - Literature Line Graph')
ax1.legend()

# Rotate x-axis labels by 45 degrees
for ax in axes:
    ax.tick_params(axis='x', labelrotation=45)

# Add annotations with spacing for each year in the same color as the lines
for ax, column in zip(axes, pivot_data.columns):
    for i, val in enumerate(pivot_data[column]):
        ax.annotate(val, (pivot_data.index[i], val), textcoords="offset points", xytext=(0, 25),
                     ha='center', fontsize=8, color=ax.get_lines()[0].get_color())

# Adjust layout and display the subplots
plt.xticks(all_years)  # Set x-ticks to include all years
plt.tight_layout()
plt.show()
```
