```python
import scipy as scipy
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
from scipy.interpolate import interp1d
```

```python
# load dataset
df = pd.read_excel('Heatmap Dataset.xlsx')
```

```python
# get valuecounts
continents = df['country name'].value_counts().to_list()
```

```python
df.columns
```

```python
m = interp1d([1,max(continents)], [7,10])
```

```python
typelist = ['open-street-map','carto-positron','stamen-terrain']
circle_radius=m(continents)
```

```python
circle_radius
```

```python
# plot heatmap on top of world map
for i in typelist:
    print(i)
    fig = px.density_mapbox(df, lat = 'latitude', lon= 'longitude', radius = circle_radius, zoom = 0, mapbox_style = i)
    fig.show()
```
