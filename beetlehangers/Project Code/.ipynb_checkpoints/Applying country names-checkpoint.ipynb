```python
import pycountry

def get_country_name(country_code):
    try:
        country = pycountry.countries.get(alpha_2=country_code)
        if country:
            return country.name
        else:
            return "Country not found"
    except LookupError:
        return "Invalid country code"

# Example usage
country_code = "us"
country_name = get_country_name(country_code)
print(country_name)
```

```python
df_source_not_inaturalist = df.loc[df['Source'] != 'iNaturalist']
```

```python
from geopy.geocoders import Nominatim

def get_country(lat_lon):
    """
    Returns the English country name based on the given latitude and longitude values.

    Args:
        lat_lon (str): Combined latitude and longitude value in the format "lat,lon".

    Returns:
        str: English country name.

    Example:
        >>> get_country("52.2337172,21.07143224")
        'Poland'
    """
    lat, lon = map(float, lat_lon.split(","))
    geolocator = Nominatim(user_agent="country_finder")
    location = geolocator.reverse((lat, lon), exactly_one=True)
    if location is None:
        return "Need to be checked"
    address = location.raw['address']
    country = address.get('country', '')
    print(country)
    if 'en' in address.get('languages', ''):
        return country
    else:
        return "Need to be checked"
```

```python
df_source_not_inaturalist['country_name'] = df_source_not_inaturalist['combined_lat_and_lon'].apply(get_country)
```
